```python
# Import necessary libraries
import pandas as pd
from datetime import datetime
from IPython.display import display, Markdown
todays_date = datetime.now().strftime("%Y-%m-%d")
student_id = "derave1577"
# Load the Titanic dataset
data = pd.read_csv('titanic.csv')
# Display the first few rows to understand what the data looks like
display(Markdown("### Step 1.2: Load the Dataset"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
display(data.head())
```

```python
display(Markdown("### Step 1.3: Initial Exploration of the Dataset"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
# Get a concise summary of the DataFrame
print(data.info())
# Get descriptive statistics that summarize the central tendency,
# dispersion, and shape of the dataset's numerical features
display(data.describe())
```

```python
display(Markdown("### Step 1.4: Data Cleaning"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
# Check for missing values in the dataset
missing_values = data.isnull().sum()
print(f"Missing values in each column:\n{missing_values}")
# Fill missing values for 'Age' with the median age
data['Age'].fillna(data['Age'].median(), inplace=True)
# Fill missing values for 'Embarked' with the most common port
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
# Drop ‘Cabin’ column due to significate number of missing values
data.drop(columns=['Cabin'], inplace=True)
# Verify that there are no remaining missing values
missing_values_after = data.isnull().sum()
print(f"Missing values after cleaning:\n{missing_values_after}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
display(Markdown("### Step 2.1: Univariate Analysis"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
# Set the aesthetic style of the plots
sns.set(style="whitegrid")
# Plot histograms for numerical features
numerical_features = ['Age', 'Fare']
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
for i, feature in enumerate(numerical_features):
    sns.histplot(data[feature], ax=axes[i], kde=True)
    axes[i].set_title(f'Histogram of {feature}')
plt.tight_layout()
plt.show()
# Plot boxplots for numerical features
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
for i, feature in enumerate(numerical_features):
    sns.boxplot(y=data[feature], ax=axes[i])
    axes[i].set_title(f'Boxplot of {feature}')
plt.tight_layout()
plt.show()
# Plot bar charts for categorical features
categorical_features = ['Pclass', 'Sex', 'SibSp', 'Parch', 'Embarked']
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 10))
axes = axes.flatten()
for i, feature in enumerate(categorical_features):
    sns.countplot(x=data[feature], ax=axes[i])
    axes[i].set_title(f'Countplot of {feature}')
plt.tight_layout()
plt.show()
```

```python
display(Markdown("### Step 2.2: Bivariate Analysis"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
# Scatter plots for numerical features against 'Survived'
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
for i, feature in enumerate(numerical_features):
    sns.scatterplot(x=data[feature], y=data['Survived'], ax=axes[i])
    axes[i].set_title(f'Scatter plot of {feature} vs Survived')
plt.tight_layout()
plt.show()
# Cross-tabulations for categorical features against 'Survived'
for feature in categorical_features:
    cross_tab = pd.crosstab(data[feature], data['Survived'],
    normalize='index')
    cross_tab.plot(kind='bar', stacked=True, figsize=(10, 5))
    plt.title(f'Survived vs {feature}')
    plt.ylabel('Proportion')
    plt.show()
```

```python
display(Markdown("### Step 3.1: Pair Plot and Heatmap"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
# Pair plot for numerical features
sns.pairplot(data[numerical_features + ['Survived']], hue='Survived', palette='viridis')
plt.suptitle('Pair Plot of Numerical Features', y=1.02)
plt.show()
# Heatmap of the correlation matrix
corr_matrix = data[['Age', 'Fare', 'Pclass', 'SibSp', 'Parch', 'Survived']].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm',
linewidths=0.5)
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
display(Markdown("### Step 3.2: Violin Plot"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
# Violin plot of Age distribution by Survival status
plt.figure(figsize=(10, 6))
sns.violinplot(x='Survived', y='Age', data=data, palette='muted', hue='Survived')
plt.title('Violin Plot of Age Distribution by Survival Status')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```

```python
display(Markdown("### Step 3.3: Box Plot"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
# Box plot of Fare distribution by Passenger Class
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Fare', data=data, palette='pastel')
plt.title('Box Plot of Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

```python
display(Markdown("### Step 3.4: Bar Plot"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
# Bar plot of survival counts by Embarked
plt.figure(figsize=(10, 6))
sns.countplot(x='Embarked', hue='Survived', data=data, palette='Set2')
plt.title('Bar Plot of Survival Counts by Embarked')
plt.xlabel('Port of Embarkation')
plt.ylabel('Count')
plt.show()
```

```python
display(Markdown("### Step 3.5: Scatter Plot"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
# Scatter plot of Age vs Fare colored by Survival status
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=data, palette='coolwarm', alpha=0.6)
plt.title('Scatter Plot of Age vs Fare Colored by Survival Status')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
display(Markdown("### Step 3.6: Line Plot"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
# Line plot of Fare over passenger index, colored by Survival status
plt.figure(figsize=(10, 6))
sns.lineplot(x=data.index, y='Fare', hue='Survived', data=data, palette='tab10')
plt.title('Line Plot of Fare Over Passenger Index Colored by Survival Status')
plt.xlabel('Passenger Index')
plt.ylabel('Fare')
plt.show()
```

```python
display(Markdown("### Step 3.7: Numerical Feature Histograms"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))
# Histograms for numerical features
data.hist(bins=10, figsize=(15, 10), layout=(4, 3))
plt.suptitle('Histograms of All Numerical Features')
plt.show()
```
